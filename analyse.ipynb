```python
import pandas as pd 
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_livre.csv')
```

```python
df['dispo'] = df['disponibilites'].str.contains('In')
df['quantite'] = df['disponibilites'].apply(lambda x: re.findall(r'\d+', x)[0])
df['quantite'] = df['quantite'].astype(int) 
df
```

```python
# Calculer la correlation entre les colonnes
corr = df.corr()

# Afficher la matrice de corrÃ©lation avec seaborn
sns.heatmap(corr, annot=True, cmap='RdYlGn', linewidths=0.2)

# Ajuster les dimensions de la figure
fig = plt.gcf()
fig.set_size_inches(6, 5)

# Afficher la figure
plt.show()
```

```python
import seaborn as sns


_ = sns.pairplot(data=df, hue="prix", palette="viridis")
```

```python
df.hist(figsize=(12, 10), bins=30, edgecolor="black")
plt.subplots_adjust(hspace=0.7, wspace=0.4)
```

```python
print(df.describe())
```

```python
data = df.sort_values(by='prix', ascending=False, inplace=False)
data10 = data.head(10)

Prix = data10['prix']
Titre = data10['titre']
# Create a bar chart
plt.barh(Titre, Prix)

# Add a title and axis labels
plt.title('les livres les plus chers ')
plt.xlabel('prix')
plt.ylabel('Les livres')

# Show the plot
plt.show()
```

```python
rating = df.sort_values(by='rating', ascending=False, inplace=False)
rating10 = rating.head(10)
Rating = rating10['rating']
Titre = rating10['titre']
# Create a bar chart
plt.barh(Titre, Rating)

# Add a title and axis labels
plt.title('les livres les mieux notÃ©s')
plt.xlabel('Ã©valuation')
plt.ylabel('Les livres')

# Show the plot
plt.show()
```
