```python
from wordcloud import WordCloud , STOPWORDS , ImageColorGenerator
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import itertools
import collections
```

```python
stopwords = set(STOPWORDS)
stopwords = list(stopwords)
stopwords.append('https')
stopwords.append('t')
stopwords.append('s')
mask = np.array(Image.open('twitter.png'))
mask[mask == 1] = 255
text = ''
```

```python
rows = []
with open('livre_csv/the_art_of_war.csv', 'r',encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)
        words = []
        for row in rows:
            for cell in row:
                words.extend(cell.split())
bigrams = list(zip(words, words[1:]))
trigrams = list(zip(words, words[1:], words[2:]))

from collections import Counter

bigram_counts = Counter(bigrams)
trigram_counts = Counter(trigrams)

num_bigrams = len(bigrams)
num_trigrams = len(trigrams)

bigram_frequencies = {bigram: count / num_bigrams for bigram, count in bigram_counts.items()}
trigram_frequencies = {trigram: count / num_trigrams for trigram, count in trigram_counts.items()}

```

```python


def get_ngrams(input_list, n):
  # Extraction des n-grammes de la liste d'entre
  return list(itertools.zip_longest(*[input_list[i:] for i in range(n)]))

rows = []
with open('livre_csv/the_art_of_war.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)
        words = []
        for row in rows:
            for cell in row:
                words.extend(cell.split())
# Calcul de la frquence des bigrammes
bigrams = get_ngrams(words, 2)
bigram_counter = collections.Counter(bigrams)
print(bigram_counter.most_common(5))

# Calcul de la frquence des trigrammes
trigrams = get_ngrams(words, 3)
trigram_counter = collections.Counter(trigrams)
xx = trigram_counter.most_common(5)
print(xx)
```

```python
from prettytable import PrettyTable

# Crez un tableau avec les colonnes "bigram" et "frequency"
table = PrettyTable(['bigram', 'xx'])

# Ajoutez chaque bigramme et sa frquence au tableau
for bigram, frequency in trigram_counter.items():
    table.add_row([trigrams, frequency])

# Affichez le tableau
print(table)
```
